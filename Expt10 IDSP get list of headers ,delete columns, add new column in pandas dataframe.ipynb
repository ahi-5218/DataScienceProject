Expt10 IDSF : Write a Python to get a list of the column headers from a Pandas DataFrame, delete DataFrame columns by name or index, add new column to existing DataFrame.

```python
import pandas as pd
```

```python
# DataFrame
data = {'Courses': ['ML', 'IDSP', 'EDAF'],
        'Fee': [2500, 3000, 2800],
        'Duration': [50, 25, 20]}
df = pd.DataFrame(data)
```

```python
# Get a list of the column headers
column_headers = list(df.columns.values)
print("The Column Header :", column_headers)
```

```python
# Delete DataFrame columns by name
df = df.drop('Duration', axis=1)
print("\nDataFrame after deleting 'City' column:\n", df)
```

```python
# Delete DataFrame columns by index
df = df.drop(df.columns[[1]], axis=1)  # Delete the first column (index 0)
print("\nDataFrame after deleting first column:\n", df)
```

```python
# Add a new column to existing DataFrame
df['Mode'] = ['Online', 'Offline', 'Hybrid']
print("\nDataFrame after adding 'Country' column:\n", df)
```
