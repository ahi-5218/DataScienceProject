Expt04 IDSP :Write a python program to demonstrate basic operation on single array and multiple arrays.

```python
import numpy as np
```

```python
# Single array operations
arr1 = np.array([1, 2, 3, 4, 5])

print("Original array:", arr1)
```

```python
# Basic operations
print("Adding 2 to each element:", arr1 + 2)
print("Multiplying each element by 3:", arr1 * 3)
print("Sum of all elements:", np.sum(arr1))
print("Mean of all elements:", np.mean(arr1))
```

```python
# Multiple array operations
arr2 = np.array([6, 7, 8, 9, 10])

print("\nArray 1:", arr1)
print("Array 2:", arr2)
```

```python
# Element-wise operations
print("Addition of arrays:", arr1 + arr2)
print("Multiplication of arrays:", arr1 * arr2)
```

```python
# Concatenation
arr3 = np.concatenate((arr1, arr2))
print("\nConcatenated array:", arr3)
```
