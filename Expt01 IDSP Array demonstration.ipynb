```python
import numpy as np
```

```python
#Array of zeros
arr_zeros = np.zeros(5)
print("Array of zeros:", arr_zeros)
```

```python
#Array of zeros
arr_zeros = np.zeros(5)
print("Array of zeros:", arr_zeros)
```

```python
#Array of ones
arr_ones = np.ones(5)
print("Array of ones:", arr_ones)
```

```python

#Array with a specific value
arr_full = np.full(5, 7)
print("Array with a specific value:", arr_full)
```

```python
#Array with a range of values
arr_arange = np.arange(0, 10, 2)
print("Array with a range of values:", arr_arange)
```

```python
#Array with linearly spaced values
arr_linspace = np.linspace(0, 1, 5)
print("Array with linearly spaced values:", arr_linspace)
```

```python
#Random array
arr_random = np.random.rand(5)
print("Random array:", arr_random)
```

```python
#Identity matrix
arr_identity = np.eye(3)
print("Identity matrix:", arr_identity)
```
