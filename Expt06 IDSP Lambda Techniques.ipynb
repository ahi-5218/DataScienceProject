Expt06 IDSP : Write a python program to demonstrate lambda technique.

```python
#lambda function to calculate the square of a number
square = lambda x: x * x
add = lambda x,y: x + y
```

```python
number = int(input("Enter a number for Squaring: "))
num1 = int(input("Enter a number1 adding Squaring: "))
num2 = int(input("Enter a number2 adding Squaring: "))
```

```python
result_square = square(number)
result_add = add(num1,num2)

print("The square of", number, "is", result_square)
print("The addition of numbers", num1,",", num2, "is", result_add)
```
