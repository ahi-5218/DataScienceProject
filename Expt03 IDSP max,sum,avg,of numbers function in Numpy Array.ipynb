Expt 02 IDSP :Write a Python function to find the max in numpy array, sum all the numbers in numpy array, find avg of no.s in numpy array.

```python
import numpy as np
```

```python
#Function defination
def array_operations(arr):
  maximum = np.max(arr)
  sum_arr = np.sum(arr)
  avg = np.mean(arr)

  return maximum, sum_arr, avg
```

```python
#calling function
my_array = np.array([1, 2, 3, 4, 5])
max_val, sum_val, avg_val = array_operations(my_array)
```

```python
print("Maximum value:", max_val)
print("Sum of elements:", sum_val)
print("Average value:", avg_val)
```
